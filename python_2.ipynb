Итак, по условию задачи нам необходимо в бесконечном цикле запрашивать у пользователя позывной (назовём его переменной name) и пароль к нему (назовём его переменной password) и проверять, есть ли информация о пользователе с такими данными в нашей системе. Таким образом, нам нужно проверить, есть ли указанный позывной среди ключей словаря secret_passwords, а затем проверить, совпадает ли введённый пароль со значением по введённому ключу.

Посмотрим на наш алгоритм создания цикла while:

Определяем условие выполнения цикла — это будет значение True, то есть цикл будет бесконечным. Условие остановки такого цикла — совпадение введённого пароля с паролем в словаре secret_passwords.

Задаём действие в цикле — запросить с помощью функции input() у пользователя его позывной, проверить, есть ли указанный позывной среди ключей словаря secret_passwords. Если позывной незнаком системе, будем выводить сообщение 'Wrong name'. Если позывной совпадает, то у пользователя запрашивается пароль.

Теперь давайте обсудим, как этот метод можно обобщить на двумерный случай — вложенный список.

Для начала запишем цикл по индексам элементов внешнего списка matrix — это будет цикл по последовательности чисел от 0 до N (не включая N), где N — длина списка matrix (в нашем случае, N = 3). Переменную цикла традиционно назовём i. В цикле будем выводить текущее значение i и значение i-го элемента списка matrix:

```python
matrix = [
    [1, 2], 
    [3, 4], 
    [5, 6]
]
N = len(matrix) #вычисляем длину внешнего списка
#создаём цикл по последовательности чисел от 0 до N (не включая N)
for i in range(N): #i — текущий элемент последовательности (индекс строки)
    print('Current i', i) #выводим текущее значение i
    print('Current row', matrix[i]) #выводим i-е значение внешнего списка
```

Теперь нам осталось организовать цикл по индексам элементов каждого из списков, то есть по индексам столбцов.

Для этого записываем вложенный цикл for по последовательности чисел от 0 до M (не включая M), где M — длина вложенного списка (в нашем случае M = 2). Переменную такого цикла по индексам столбцов традиционно называют j. Во внутреннем цикле будет выводить текущее значение j и значение элемента списка matrix под индексами i-й строки и j-го столбца.

Условие задачи. Нам предоставлены данные о баллах пяти студентов по трём экзаменам. Данные находятся во вложенном списке student_scores.

student_scores = [
  
    [56, 90, 80],
    [80, 86, 92],
    [91, 76, 89],
    [91, 42, 60],
    [65, 30, 90]
]
Каждая строка таблицы соответствует номеру студента, а каждый из трёх столбцов соответствует номеру экзамена (первый — математика, второй — информатика, третий — русский язык).

Необходимо найти средний балл студентов по каждому из экзаменов и общий средний балл по всем экзаменам.

Очевидно, что средний балл по каждому из экзаменов рассчитывается как:

\frac{общий\ балл\ по\ экзамену}{количество\ студентов}

Общий средний балл рассчитывается так:

\frac{общий\ балл\ по\ всем\ экзаменам}{количество\ студентов\ *\ количество\ экзаменов}

Количество студентов нам известно — это длина списка student_scores, и она равна 5. Количество экзаменов нам тоже известно — это длина любого из внутренних списков, и она равна 3.

Осталось только посчитать общий балл по всем экзаменам и общий балл по математике, информатике и русскому языку. Для этого создадим четыре соответствующие переменные: summa, math_sum, info_sum и rus_sum, в которых будем считать суммы. В цикле будем увеличивать их значения.

Для прохода по списку с баллами будем использовать вложенный цикл по его индексам. 

Внешний цикл — цикл по последовательности от 0 до N (не включая N), где N — число студентов (цикл по студентам). Переменную внешнего цикла назовём i.

Внутренний цикл — цикл по последовательности от 0 до M (не включая M), где M — число экзаменов (цикл по экзаменам). Переменную внутреннего цикла назовем j


Во внешнем цикле будем считать сумму баллов студентов по каждому из предметов. Так как баллы по математике находятся в первом столбце таблицы, нам нужно к переменной math_sum прибавить значение из i-й строки и столбца под индексом 0. Аналогично увеличиваем переменные info_sum и rus_sum, меняя только номер столбца.

Во внутреннем цикле будем считать общую сумму баллов студентов по всем предметам. Для этого в цикле необходимо увеличивать переменную summa на значение из i-й строки и j-го столбца таблицы (складываем все числа в таблице).

По окончании цикла выводим средние баллы по предметам и общий средний балл.

Наш итоговый код будет иметь следующий вид:

```python
student_scores = [
    [56, 90, 80],
    [80, 86, 92],
    [91, 76, 89],
    [91, 42, 60],
    [65, 30, 90]
] #заданная таблица со студентами
N = len(student_scores) #задаём число студентов
M = len(student_scores[0]) #задаём число экзаменов
summa = 0 #задаём начальное значение общего балла
math_sum = 0 #задаём начальное значение общего балла по математике
info_sum = 0 #задаём начальное значение общего балла по информатике
rus_sum = 0 #задаём начальное значение общего балла по русскому языку
#создаём цикл по последовательности от 0 до N (не включая N)
for i in range(N): #i — индекс строки
    math_sum += student_scores[i][0] #добавляем баллы по математике i-го студента
    info_sum += student_scores[i][1] #добавляем баллы по информатике i-го студента
    rus_sum += student_scores[i][2] #добавляем баллы по русскому i-го студента
    #создаём цикл по последовательности от 0 до M (не включая M)
    for j in range(M): #j — индекс столбца
        summa += student_scores[i][j] #добавляем баллы i-го студента по j-му экзамену
print('Average math score', math_sum/N) #выводим средний балл по математике
print('Average info score', info_sum/N) #выводим средний балл по информатике
print('Average rus score', rus_sum/N) #выводим средний балл по русскому языку
print('Average score', summa/(N*M)) #выводим общий средний балл
```

```python
table = [
    [1, 3, 6],
    [4, 6, 8],
    [10, 33, 53]
]
N = len(table) #вычисляем количество строк
M = len(table[0]) #вычисляем количество столбцов
S = 0 #задаём переменную суммы
count = 0 #задаём переменную количества итераций
#создаём цикл по индексам строк таблицы
for i in range(N): # i — индекс строки
    #создаём цикл по индексам столбцов таблицы
    for j in range(M): #j — индекс столбца
        #проверяем условие, что индекс столбца — 0
        if j == 0: #если условие истинно,
            S += table[i][j] #увеличиваем сумму
        count += 1 #увеличиваем количество итераций
print('Summa', S) #выводим сумму 
print('Count iteration', count) #выводим количество итераций
```
